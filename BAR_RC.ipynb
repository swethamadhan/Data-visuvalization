#BAR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#####Simple bar plot:


```python
fig = plt.figure(figsize = (10, 5))
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())
plt.bar(courses, values, color ='green',width = 0.5)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

#####Unstacked Bar:

```python
plotdata = pd.DataFrame({"2018":[57,67,77,83],
 "2019":[68,73,80,79],
 "2020":[73,78,80,85]},
index=["Django", "Gafur", "Tommy", "Ronnie"])
plotdata.plot(kind="bar",figsize=(12, 6))
plt.title("FIFA ratings")
plt.xlabel("Footballer")
plt.ylabel("Ratings")
```

#####Stacked bar plots:


```python
plotdata = pd.DataFrame({"2018":[57,67,77,83],
"2019":[68,73,80,79],
 "2020":[73,78,80,85]},
index=["Django", "Gafur", "Tommy", "Ronnie"])
plotdata.plot(kind='bar', stacked=True,figsize=(10, 5),color=['b','g','r'])
plt.title("FIFA ratings")
plt.xlabel("Footballer")
plt.ylabel("Ratings")
plt.show()
```

#RC

```python
import scipy.stats
 
x = [15,18,21, 15, 21]
y = [25,25,27,27,27]
 
def spearmans_rank_correlation(x, y):
    xranks = pd.Series(x).rank()
    print("Rankings of X:")
    print(xranks)
    yranks = pd.Series(y).rank()
    print("Rankings of Y:")
    print(yranks)
    print("Spearman's Rank correlation:", scipy.stats.pearsonr(xranks, yranks)[0]) 
spearmans_rank_correlation(x, y)
```
